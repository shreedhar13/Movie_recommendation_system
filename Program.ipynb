```python
#above u see ' [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}] '
#it is a string.,, we have to convert it to list for further evaluation
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
#string to iist ,,converted..
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()#successfully genres column is updated
```

```python
#same for keyword
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#in cast column i just want top 5 character names,,,,bcz they are so famous thats why mentioned in top

def convert5(obj):
    count = 0
    L=[]
    for i in ast.literal_eval(obj):
        if count != 5:
            L.append(i['name'])
            count +=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
#user see a movie of a particular director , writer more and more ,,,so extract director and writer of movie..
#bcz some rows have jjust director,so i use director only
def fetch_director_writer(obj):
    
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name']) 
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director_writer)
```

```python
movies.head()
```

```python
movies['crew']
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#as u see Sam Worthington  and Sam Mendes are two diff things,,,but while creating tags for content based recommendation system,,it will take sam as 1 word then worthington is another word..etc
#so to avoid it we have to combine space separated words
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head()#as u see ,science fiction become sciencefiction...not we can identify each word uniquel...
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#Now we have  data  which is in the write format to achieve our goal
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head()#required form of dataset..
```

```python
#del movies    #to release memory
```

```python
#removing list form of tags columns
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#Do stemming..
ps=PorterStemmer()
def stemming(text):
    t=[]
    
    for i in text.split():
        t.append(ps.stem(i))
    return " ".join(t)
    
```

```python
new_df['tags'] = new_df['tags'].apply(stemming)
```

```python
new_df['tags'][0]#stemming is done successfully
```

```python
#Lowercasing all words of tags column
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
new_df.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')#remove all stop words from english language ,,present in corpus..
#max_features=5000 -> consider top 5k words from corpus which are more common in corpus,,ie;frequency of those words is more..
```

```python
# above all 3 steps is done by this CountVectorizer class internally/automatically
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
#cv.fit_transform returns scifi sparse matrix object..so convert it to array
```

# Distance or angle inversally proposnal to similarity

```python
#cosine_distance -> larger the distance more dissimilar movies,,,,smaller the distance more similar movies
#cosine_similarity _(ranges from 0 to 1),,,,nearer ro 1->means high similarity b/n movies,,,nearer to 0->means less similar
```

# similarity inversaly propostional to Distance

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix = cosine_similarity(vectors)
#there are total 4806 vectors (movies) ,each with 5000 dimension..
#ie;calculate similarity b/n each and every vector
#ie;first movie(1st vector) is compared with all 4806 vectors ie;calculating cosine_similarity...like this for each and every movie we will calculate this.. 

#Thus we get 4806 x 4806  array....similarity value matrix..
#4806x4806 = 23,097,636 these many calculation is done..
```

```python
similarity_matrix.shape
```

```python
similarity_matrix
```

```python
similarity_matrix[0].tolist()#similarity of first movie with all other mivies including itsef (4806)
#thats why similarity of first movie with itself is 1,,,ie;around 0.99999999999999998 ~ 1
#similarity score of first movie with second movie is 0.06818181818181818 ,,,so on...order matters
```

```python
#thus diagonal of similarity_matrix is 1 or 0.9999999999999999
```

# RECOMMENDATION SYSTEM LOGIC

```python
sorted(  list( enumerate( similarity_matrix[0] ) ) , reverse=True , key=lambda x:x[1])[0:10] #recommend top 10 movie
#key=lambda x:x[1]  -> sort based on element present at 1th index ....
#list of tuples...[(oth index , 1th index),....]
#oth index -> index of movie ,, 1th index -> similarity value with 1st movie(1st movie bcz we have choosen first movie similarity_matrix[0] )
```

```python
# but i dont want same movie,,bcz each movie is exactly similar with itself (0.99999999),,so dont consider it
sorted(  list( enumerate( similarity_matrix[0] ) ) , reverse=True , key=lambda x:x[1])[1:11]#11-1 => 10,,,list slicing
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie ].index[0]#fetching index of searched movie,,or movie on which we want to perform recommedation algo
    similarity = similarity_matrix[movie_index]
    movies_list = sorted(  list( enumerate( similarity ) ) , reverse=True , key=lambda x:x[1])[1:11]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')#as u see Avatar is a alien and universe based movie,,so our content based recommedation system  recommending us Alien related movies..
#see ,, out system captured similarity more accurately
```
